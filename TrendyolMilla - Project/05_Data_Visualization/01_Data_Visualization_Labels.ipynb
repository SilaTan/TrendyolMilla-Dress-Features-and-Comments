```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import os
import matplotlib.ticker as ticker
```

```python
pd.set_option('display.max_colwidth', None) 
pd.set_option('display.max_rows', None) 
pd.set_option('display.max_columns', None)  
```

```python
comments_path = os.path.join(os.path.expanduser("~"), "Desktop\\TrendyolMilla\\03_Data_Labelization\\Data", "comments_final_labels.csv") 

df_comments = pd.read_csv(comments_path)

df_comments.head()
```

Label - 1 : savasy_label

```python
label_counts = df_comments["savasy_label"].value_counts()

plt.figure(figsize=(8, 5))
label_counts.plot(kind="bar", color=["red", "green"])  
plt.xlabel("Duygu Etiketi")
plt.ylabel("Frekans")
plt.title("Savasy Modeli - Yorum Duygu Dağılımı")
plt.xticks(rotation=0)  
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df_comments["savasy_skor"], bins=30, color='orange', edgecolor='black')
plt.title("Savasy_skor Dağılımı")
plt.xlabel("Değer")
plt.ylabel("Frekans")
plt.grid(True)
plt.show()
```
