Yorum DataFrame İçeri Aktarım

```python
comments_folder = os.path.join(os.path.expanduser("~"), "Desktop", "TrendyolMilla","01_Data_Extraction" ,"Data", "Comments")


df_list = []


for i in range(10, 1980, 10):   
    file_name = f"comments_{i}.csv"
    file_path = os.path.join(comments_folder, file_name)


    if os.path.exists(file_path):
        df = pd.read_csv(file_path)   
        df_list.append(df)  
        print(f"Okundu: {file_name}")  


df_comments = pd.concat(df_list, ignore_index=True)


df_comments.head()
```
